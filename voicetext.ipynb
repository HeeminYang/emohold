```python
# 기존 listdir + MacOS 시스템으로 인한 (더미 파일(.DS_Store) 제거 + 한글 인식 모듈 변경) 함수
def listdir_ds_remover(dir):
    import os
    from unicodedata import normalize
    if os.path.isfile(dir+'/.DS_Store'):
        os.remove(dir+'/.DS_Store')
        list = os.listdir(dir)
        new_list=[]
        for i in list:
            i = normalize("NFC", i)
            new_list.append(i)
        return new_list
    else :
        list = os.listdir(dir)
        new_list=[]
        for i in list:
            i = normalize("NFC", i)
            new_list.append(i)
        return new_list
```

```python
# Import the necessary modules
from transformers import BertModel, DistilBertModel, BertForSequenceClassification, DistilBertForSequenceClassification
from tokenization_kobert import KoBertTokenizer
import pandas as pd
import torch

# Load the KoBERT model
# bert_model = BertModel.from_pretrained('monologg/kobert')
# distilbert_model = DistilBertModel.from_pretrained('monologg/distilkobert')

tokenizer = KoBertTokenizer.from_pretrained('monologg/kobert') # monologg/distilkobert도 동일

# tokenizer = AutoTokenizer.from_pretrained("monologg/kobert")
bert_model = BertForSequenceClassification.from_pretrained("monologg/kobert", problem_type="multi_label_classification")
distilbert_model = DistilBertForSequenceClassification.from_pretrained("monologg/distilkobert", problem_type="multi_label_classification")

for file in listdir_ds_remover('/home/heemin/mv/dir/voice_csv'):
    dataframe = pd.read_csv('/home/heemin/mv/dir/voice_csv/'+file)

    # Get the "speech" column data from the dataframe
    speech_data = dataframe["Speech"]

    # Add a new "sentiment" column to the dataframe
    # dataframe["sentiment"] = ""

    # Loop through the "speech" data and perform sentiment analysis on each sentence
    for i, speech in enumerate(speech_data):
        # Tokenize the sentence
        tokens = tokenizer.tokenize(speech)

        # Convert the tokens to their corresponding index numbers
        input_ids = tokenizer.convert_tokens_to_ids(tokens)

        # Format the input for the neural network
        inputs = torch.tensor([input_ids])
        bert_model.eval()
        # Perform sentiment analysis
        outputs1 = bert_model(inputs)
        outputs2 = distilbert_model(inputs)

        # Get the predicted sentiment class
        predicted_class1 = outputs1[0].argmax().item()
        predicted_class2 = outputs2[0].argmax().item()


        # Store the predicted sentiment in the "sentiment" column of the dataframe
        dataframe.loc[i, "kobert"] = predicted_class1
        dataframe.loc[i, "distilkobert"] = predicted_class2

        # if predicted_class == 0:
        #     dataframe.loc[i, "sentiment"] = "positive"
        # elif predicted_class == 1:
        #     dataframe.loc[i, "sentiment"] = "negative"
        # else:
        #     dataframe.loc[i, "sentiment"] = "neutral"

    dataframe.to_csv('/home/heemin/mv/dir/voice_distilkobert/'+file)
```
