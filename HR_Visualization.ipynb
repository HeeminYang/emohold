### <span style="font-size:180%"> Introduce Simple Function </span>
> <span style="font-size:180%"><span style="color:Springgreen">HR_todf(<span style="color:Khaki">file_path</span>)</span>: HR json 파일을 dataframe으로 불러옴 </span>

> <span style="font-size:180%"><span style="color:Springgreen">HR_time_selection(<span style="color:Khaki">HR_todf(), date, start_time, end_time</span>)</span>:   
<span style="color:Khaki">HR_todf()</span>으로 불러온 dataframe을 <span style="color:Khaki">date</span>(날짜)를 지정하고 <span style="color:Khaki">start, end time</span>으로 잘라줌 </span>

> <span style="font-size:180%"><span style="color:Springgreen">HR_makeplot(<span style="color:Khaki">id_name, date, HR_time_selection(), line1, line2, img_path, id</span>)</span>:

<span style="color:Khaki">line1, line2</span>로 교육 시작과 끝을 표시함,   
<span style="color:Khaki">img_path</span>는 <span style="color:LimeGreen">code</span> 폴더 밖에 폴더이름을 입력받음   
<span style="color:Khaki">id</span>는 이미지 개별 저장을 위한 구분인자</span></span>
