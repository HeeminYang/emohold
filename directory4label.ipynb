```python
from glob import glob
```

```python
filenames = glob('/home/projects/Metaverse2021/Crop_Emotion/*/*/*/*/*')
filenames[:5]
```

```python
print(filenames[0].split('/'))
```

```python
def get_info(path):
    path_split = path.split('/')
    superior = '/'.join(path_split[:5])
    grade = path_split[5]
    name = path_split[6]
    session = path_split[7]
    emotion = path_split[8]
    filename = path_split[9]

    return pd.Series({'path':path,'superior' : superior,'grade':grade, 'name':name, 'session': session, 'emotion': emotion, 'filename':filename})
```

```python
temp = pd.Series(filenames)
df = temp.apply(get_info)

df
```

```python
df['new_filename'] = 1
df['new_filename'] = df.groupby('name')['new_filename'].apply(np.cumsum)
```

```python
df['new_filename'] = df['new_filename']-1
```

```python
df[df['name']=='KSY']
```

```python
temp = df['superior'] + '/' + df['name'] +'_'+ df['new_filename'].astype(str)+'.mp4'
temp[0]
```

```python
path_df = pd.DataFrame()

path_df['old_path'] = df['path']
path_df['new_path'] = temp
path_df['file'] = temp.str.split('/').str[-1].str.replace('.mp4','')
path_df['session'] = df['session']
path_df['emotion'] = df['emotion']
```

```python
path_df
```

```python
import shutil
for i,rows in path_df.iterrows():
    shutil.move(rows['old_path'], rows['new_path'])
```
